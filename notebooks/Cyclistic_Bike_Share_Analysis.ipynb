```python
import numpy as np
import pandas as pd
import glob
```

```python
df1 = pd.read_csv("202504-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df2 = pd.read_csv("202505-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df3 = pd.read_csv("202506-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df4 = pd.read_csv("202507-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df5 = pd.read_csv("202508-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df6 = pd.read_csv("202509-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df7 = pd.read_csv("202510-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df8 = pd.read_csv("202511-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df9 = pd.read_csv("202512-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df10 = pd.read_csv("202601-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df11 = pd.read_csv("202602-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
df12 = pd.read_csv("202603-divvy-tripdata.csv", engine="python", on_bad_lines="skip")
```

```python
df = pd.concat(
    [df1, df2, df3, df4, df5, df6,
     df7, df8, df9, df10, df11, df12],
    ignore_index=True
)
```

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
df[['started_at', 'ended_at']].dtypes
```

```python
df['ride_length'] = (
    df['ended_at'] - df['started_at']
).dt.total_seconds() / 60
```

```python
df['ride_length'].describe()
```

```python
(df['ride_length'] > 60).sum()
```

```python
(df['ride_length'] > 1440).sum()
```

```python
(df['ride_length'] < 0).sum()
```

```python
df = df[df['ride_length'] >0 ]
```

```python
df = df[df['ride_length'] <= 1440]
```

```python
df['day_of_week'] = df['started_at'].dt.day_name()

df['month'] = df['started_at'].dt.month_name()

df['hour'] = df['started_at'].dt.hour
```
